```python
import os
import pandas as pd
```

```python
# Setting variables to be used in format string command below
#tweet_count = 500
text_query = "Zuckerberg"
since_date = "2021-10-28"
until_date = "2021-11-05"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --since {} twitter-search "{} until:{}"> zuckerberg.json'
          .format(since_date, text_query, until_date))
```

```python
df = pd.read_json('zuckerberg.json', lines = True)
```
